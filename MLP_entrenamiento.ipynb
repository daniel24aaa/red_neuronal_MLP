```python
# armando el data set
centroides = centroides_normalizado

relaciones = relaciones_normalizado

datos_dataset = []

for i in range(len(centroides)):
    datos_imagen = {
        'centroides': centroides[i],
        'relaciones': relaciones[i]
    }
    datos_dataset.append(datos_imagen)
```

```python
#entrenamiento con 1 capa oculta
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt


# Separar los datos en conjuntos de entrenamiento y prueba
train_data, test_data = train_test_split(datos_dataset, test_size=0.2, random_state=42)

# Preparar los datos para el modelo
def prepare_data(data):
    inputs = [torch.FloatTensor(imagen['centroides']).view(-1) for imagen in data]
    targets = torch.FloatTensor([imagen['relaciones'] for imagen in data])
    return list(zip(inputs, targets))

train_data_prepared = prepare_data(train_data)
test_data_prepared = prepare_data(test_data)

# Definir el modelo MLP
class MLPModel(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(MLPModel, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

# Tamao de entrada, oculta y salida (ajustar segn tu problema)
input_size = len(train_data_prepared[0][0])  # Tamao de la entrada (nmero de caractersticas)
hidden_size = 16 #16
output_size = len(train_data_prepared[0][1])  # Tamao de la salida (nmero de relaciones)

# Crear el modelo
model = MLPModel(input_size, hidden_size, output_size)

# Funcin de prdida y optimizador
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)



# Funcin para entrenar el modelo y devolver una lista con las prdidas
def train_model(model, train_data, criterion, optimizer, num_epochs=1000):
    losses = []
    for epoch in range(num_epochs):
        total_loss = 0
        for inputs, targets in train_data:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        losses.append(total_loss)

        if (epoch+1) % 100 == 0:
            print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss:.4f}')

    return losses

# Entrenar el modelo y obtener las prdidas
losses = train_model(model, train_data_prepared, criterion, optimizer, num_epochs=3000)

# Graficar las prdidas
plt.plot(losses)
plt.title('Evolucin de la Prdida durante el Entrenamiento')
plt.xlabel('poca')
plt.ylabel('Prdida')
plt.show()

# Evaluar el modelo en el conjunto de prueba
model.eval()
with torch.no_grad():
    total_loss = 0
    for inputs, targets in test_data_prepared:
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        total_loss += loss.item()

    print(f'Mean Squared Error on Test Data: {total_loss:.4f}')
```

```python
#modelo con 2 capas ocultas
import torch
import torch.nn as nn
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
    

# Separar los datos en conjuntos de entrenamiento y prueba
train_data, test_data = train_test_split(datos_dataset, test_size=0.2, random_state=42)

# Preparar los datos para el modelo
def prepare_data(data):
    inputs = [torch.FloatTensor(imagen['centroides']).view(-1) for imagen in data]
    targets = torch.FloatTensor([imagen['relaciones'] for imagen in data])
    return list(zip(inputs, targets))

train_data_prepared = prepare_data(train_data)
test_data_prepared = prepare_data(test_data)

# Definir el modelo MLP
class MLPModel(nn.Module):
    def __init__(self, input_size, hidden_size1, hidden_size2, output_size):
        super(MLPModel, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size1)
        self.relu1 = nn.ReLU()
        self.fc2 = nn.Linear(hidden_size1, hidden_size2)
        self.relu2 = nn.ReLU()
        self.fc3 = nn.Linear(hidden_size2, output_size)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu1(x)
        x = self.fc2(x)
        x = self.relu2(x)
        x = self.fc3(x)
        return x

# Tamao de entrada, oculta y salida (ajustar segn tu problema)
# Ajustar el tamao de entrada, oculta1, oculta2 y salida
input_size = len(train_data_prepared[0][0])  
hidden_size1 = 8
hidden_size2 = 16
output_size = len(train_data_prepared[0][1])

# Crear el modelo con la nueva arquitectura
model = MLPModel(input_size, hidden_size1, hidden_size2, output_size)

# Funcin de prdida y optimizador
criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)



# Funcin para entrenar el modelo y devolver una lista con las prdidas
def train_model(model, train_data, criterion, optimizer, num_epochs=1000):
    losses = []
    for epoch in range(num_epochs):
        total_loss = 0
        for inputs, targets in train_data:
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, targets)
            loss.backward()
            optimizer.step()
            total_loss += loss.item()

        losses.append(total_loss)

        if (epoch+1) % 100 == 0:
            print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss:.4f}')

    return losses

# Entrenar el modelo y obtener las prdidas
losses = train_model(model, train_data_prepared, criterion, optimizer, num_epochs=1000)

# Graficar las prdidas
plt.plot(losses)
plt.title('Evolucin de la Prdida durante el Entrenamiento')
plt.xlabel('poca')
plt.ylabel('Prdida')
plt.show()

# Evaluar el modelo en el conjunto de prueba
model.eval()
with torch.no_grad():
    total_loss = 0
    for inputs, targets in test_data_prepared:
        outputs = model(inputs)
        loss = criterion(outputs, targets)
        total_loss += loss.item()

    print(f'Mean Squared Error on Test Data: {total_loss:.4f}')
```
